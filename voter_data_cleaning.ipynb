```python
column_set_a = [
    'V200001',  # Case ID
    'V201005',  # How often does respondent pay attention to politics
    'V201006',  # Interest in following campaigns
    'V201020',  # Voted in a primary/caucus
    'V201028',  # Voted in the presidential election
    'V201029',  # Candidate voted for in presidential election
    'V201018',  # Party registration
    'V201004',  # Gender
    'V201016',  # How long respondent has been registered to vote
    'V201024',  # How the respondent voted (e.g., in-person, mail)
    'V201030',  # Strength of preference for candidate voted for
    'V201222',  # Does the respondent see voting as a duty or choice
    'V202014',  # Participation in rallies or political events
    'V202016'  # Donated to a campaign
]

column_set_b = [
    'V202022',  # Discussed politics with family/friends
    'V202029',  # Posted political content online
    'V202219',  # Belief in fairness of vote counting
    'V201357',  # Opinion on requiring voter ID
    'V202426',  # Should the governemnt be concerned with income inequality
    'V201357',  # Should individuals have to show photo ID when they vote
    'V201360',  # Should felons be able to vote after finishing sentence
    'V201336',  # Should abortion be permitted
    'V202221',  # How important is it that more African Americans are elected to office
    'V202222',  # How important is it that more Asians are elected to office
    'V202223',  # How important is it that more LGBT+  are elected to office
    'V202224'  # How important is it that more Women people are elected to office
]

relevant_columns = column_set_a + column_set_b

df = df[relevant_columns]
```

There are a few concerning columns:

* 'V201028',   Voted in the presidential election
* 'V201029',  # Candidate voted for in presidential election
* 'V201004',  # Gender
* 'V201024',  # How the respondent voted (e.g., in-person, mail)
* 'V201030',  # Strength of preference for candidate voted for

Most of the candidates have not voted, so the majority of these values are null. Lets remove them.


```python
columns_to_remove = [
    'V201028',
    'V201029',
    'V201004',
    'V201024',
    'V201030'
]

df = df.drop(columns=columns_to_remove)
df.head()
```

```python
print(f"The dataframe has {df.shape[0]} rows and {df.shape[1]} columns")
```
