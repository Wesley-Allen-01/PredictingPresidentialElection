```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('data/voter_data_cleaned.csv')
df = df.drop(columns=['V200001'])
df.head()
```

```python
print(list(df.columns))
```

```python
df_onehot = pd.get_dummies(df, columns=df.columns)
df_onehot.head()
```

```python
print(list(df_onehot.columns))
```

```python
s = list(df_onehot.columns)
s = sorted(s)
print(s)
```

```python
df_onehot = df_onehot.drop(columns=['V201357.1_-8', 'V201357.1_-9', 'V201357.1_1', 'V201357.1_2', 'V201357.1_3'])
```

```python
frequent_itemsets = apriori(df_onehot, min_support=0.1, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

print(rules)
```

```python
rules.head()
```

```python
top_support = rules['support'].nlargest(15)
display(top_support)
```

```python
top_support_df = rules.loc[top_support.index]
display(top_support_df)
```

```python
top_confidence = rules['confidence'].nlargest(15)
top_confidence_df = rules.loc[top_confidence.index]
display(top_confidence_df)
```

```python
top_lift = rules['lift'].nlargest(15)
top_lift_df = rules.loc[top_lift.index]
display(top_lift_df)
```

```python
column_questions = {
    'V200001': 'Case ID',
    'V201005': 'Attention to politics',
    'V201006': 'Interest in campaigns',
    'V201020': 'Voted in primary/caucus',
    'V201028': 'Voted in presidential election',
    'V201029': 'Presidential candidate voted for',
    'V201018': 'Party registration',
    'V201004': 'Gender',
    'V201016': 'Length of voter registration',
    'V201024': 'Method of voting',
    'V201030': 'Strength of candidate preference',
    'V201222': 'Voting duty vs. choice',
    'V202014': 'Participation in rallies',
    'V202016': 'Donated to campaign',
    'V202022': 'Discussed politics with family/friends',
    'V202029': 'Posted political content online',
    'V202219': 'Belief in fair vote counting',
    'V201357': 'Opinion on voter ID',
    'V202426': 'Concern for income inequality',
    'V201360': 'Should felons vote after sentence',
    'V201336': 'Should abortion be permitted',
    'V202221': 'More African Americans in office',
    'V202222': 'More Asians in office',
    'V202223': 'More LGBT+ in office',
    'V202224': 'More women in office'
}

def rule_interpreter(frame):
    for index, row in frame.iterrows():
        antecedents = list(row['antecedents'])
        ant_codes = [str(ant).split('_')[0] for ant in antecedents]
        ant_questions = [column_questions.get(code, f"Unknown question for code {code}") for code in ant_codes]
        ant_answers = [str(ant).split('_')[1] for ant in antecedents]
        
        consequents = list(row['consequents'])
        cons_codes = [str(cons).split('_')[0] for cons in consequents]
        cons_questions = [column_questions.get(code, f"Unknown question for code {code}") for code in cons_codes]
        cons_answers = [str(cons).split('_')[1] for cons in consequents]
        
        print(f"Rule {index + 1}:")
        print("If:")
        for idx, (q, a) in enumerate(zip(ant_questions, ant_answers)):
            print(f"- {q} with answer {a}")
            print(f"\t- {ant_codes[idx]}")
        
        print("Then:")
        for idx, (q, a) in enumerate(zip(cons_questions, cons_answers)):
            print(f"- {q} with answer {a}")
            print(f"\t- {cons_codes[idx]}")
        print(f"Support: {row['support']:.4f}")
        print(f"Confidence: {row['confidence']:.4f}")
        print(f"Lift: {row['lift']:.4f}")
        print("\n")

rule_interpreter(top_support_df)
```

```python
rule_interpreter(top_confidence_df)
```

```python
rule_interpreter(top_lift_df)
```
